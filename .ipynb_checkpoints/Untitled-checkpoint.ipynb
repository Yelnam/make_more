```python
N = torch.zeros((27,27), dtype = torch.int32)
```

```python
chars = sorted(list(set(''.join(words))))
s_to_i = {s:i+1 for i, s in enumerate(chars)}
s_to_i['.'] = 0
s_to_i
```

```python
i_to_s = {i:s for s,i in s_to_i.items()}
i_to_s
```

```python
for word in words:
    tok = ['.'] + list(word) + ['.']
    for ch, ch2 in zip(tok, tok[1:]):
        ch_i = s_to_i[ch]
        ch2_i = s_to_i[ch2]
        N[ch_i, ch2_i] += 1
```

```python
# seed to make random generations deterministic and repeatable
g = torch.Generator().manual_seed(2147483647)
p = torch.rand(3, generator = g)
p = p / p.sum()
p
```

```python
# multinomial based on above 'randomly' generated probabilities
torch.multinomial(p, num_samples = 20, replacement = True, generator = g)
```

```python
for i in range(15):
    out = []
    ix = 0
    while True:
        p = N[ix].float() #turn values in row ix (starting at row 0) into floats in order to perform calculations on them
        p = p /p.sum() # normalise row to produce probabilities summing to 1
        ix = torch.multinomial(p, num_samples = 1, replacement = True, generator = g).item() # get new ix value from a single sample on the row
        out.append(i_to_s[ix])
        if ix == 0:
            break
    print(''.join(out))
```

```python
# generate matrix of probabilities across rows (avoids inefficiency of renorming every run)
P = N.float()
row_sums = P.sum(1, keepdim = True) # 1 sums across rows (0 would sum columns)
# P = P / row_sums # divide the values in each cell of a row, by the row total (normalise the row)
P /= row_sums # commented out above, run /= to perform operation in place and avoid adding new memory
```

```python
P[26].sum()
```

```python
# run again, substituting the simpler P calculation
for i in range(15):
    out = []
    ix = 0
    while True:
        p = P[ix]
        ix = torch.multinomial(p, num_samples = 1, replacement = True, generator = g).item() # get new ix value from a single sample on the row
        out.append(i_to_s[ix])
        if ix == 0:
            break
    print(''.join(out))
```

```python
# run bigrams on early names again. this time include probabilities
log_likelihood = 0.0
n = 0
for word in words[:20]:
    tok = ['.'] + list(word) + ['.']
    for ch, ch2 in zip(tok, tok[1:]):
        ch_i = s_to_i[ch]
        ch2_i = s_to_i[ch2]
        prob = P[ch_i, ch2_i]
        logprob = torch.log(prob)
        log_likelihood += logprob
        n += 1
        print(f'{ch}{ch2} \t%Prob: {prob*100:.2f} \tIndex: {(prob/(1/27)):.2f} \tLogProb: {logprob:.2f}')

nll = -log_likelihood
print(f'\nLog likelihood: {log_likelihood:.2f}, Neg log likelihood: {nll:.2f}, Ave. NLL: {nll/n:.7f}')
```

```python
# create a training set of bigrams
xs, ys = [], []
for word in words[:2]:
    tok = ['.'] + list(word) + ['.']
    for ch, ch2 in zip(tok, tok[1:]):
        ch_i = s_to_i[ch]
        ch2_i = s_to_i[ch2]
        print(ch, ch2)
        xs.append(ch_i)
        ys.append(ch2_i)
        
xs = torch.tensor(xs) # using .tensor (lower case t), which infers dtype
ys = torch.tensor(ys) #     .Tensor (upper case) would force float dtypes
```

```python
xs, ys
```

```python
# encode xs with one hot encoding
x_enc = F.one_hot(xs, num_classes = 27).float() # ensure you end up with floats (function doesn't contain a dtype arg)
x_enc
```

```python
plt.imshow(x_enc)
```

```python
# digression - personal test 
nm = ''.join("robert.manley")
nm_is = [s_to_i[s] for s in nm]
plt.imshow(F.one_hot(torch.tensor(nm_is), num_classes = 27))
```

```python
# initialise weights for NN
W = torch.randn((27,1))

# @ is a matrix multiplication in torch
x_enc @ W 

# NB in the below each value is the dot product of W with the respective row from the grid generated up above by x_enc
# ie the 0th value is each of the 27 values in the top row of the grid, * each pair-wise value in W, summed to give the DP
# because one hot gives a single 1 and otherwise 0s, each value in x_enc @ W will be:
# one times the index in W corresponding to the index where the 1 appeared in x_enc (i0 for '.', i5 for 'e' etc)
# essentially the output gives the random weight from W at the index matching the index of the letter in s_to_i
```

```python
# same again, but this time change the second dimension in W to 27
# (this gives 27 neurons. not sure (yet) why Karpathy has it match the number of characters)
# (Karpathy next explains that each neuron will output probabilities for the next character after its char - so 27)
W = torch.randn((27,27))

# @ is again a matrix multiplication in torch
x_enc @ W 

# essentially, this is the same as the above, repeated 27 times with random weights each time
```

```python
# what do the floats in xenc @ W represent?
# Karpathy's explanation:
# the 'firing rate' for a given input, over a given neuron
# e.g. to see the firing rate of the 0th input on the 0th neuron:
(x_enc @ W)[0][0]
```

```python
# Rob's explanation
# each 'row' of the tensor represents a single character
# each value in the row is the weight applied to the character by each of the 27 neurons
# i.e. the value at [0][0] is the first character '.' multiplied by the respective i from its index in W
# [1][1] is the value from [5][1] in W (character 'e' gives 5, position of neuron gives 1)
(x_enc @ W)[1][1]
```

```python
W
```

```python
# exponentiate numbers in matrix multiply result
logits = x_enc @ W
counts = logits.exp()

# normalise rows to get probabilities
probs = counts / counts.sum(1, keepdims = True)
probs

# NB SOFTMAX
# the rows counts and probs above, taken together, are essentially identical to softmax function
```

```python
#test random rows sum to 1
import random
n = 0
while n < 3:
    i = random.choice(range(len(probs)))
    print(i, probs[i].sum())
    n += 1
```

```python
# example by Karpathy using printouts to explain what is happening here
nlls = torch.zeros(5)
print('Printout of random probabilites assigned to labels (next letters in bigrams), summarised with Loss:')
for i in range(5):
    x = xs[i].item()
    y = ys[i].item()
    print(f'\n--\n\nBigram example {1}: {i_to_s[x]}{i_to_s[y]}\tIndexes {x} and {y}')
    print(f'\nInput {x} gives probabilities: \n{probs[i]}')
    print(f'\nLabel: {y}')
    p = probs[i, y]
    print(f'Probability assigned to label: {p.item()}')
    logp, nll = torch.log(p), torch.log(p) * -1
    print(f'\nLog likelihood: {logp.item()}\tNeg log likelihood: {nll.item()}')
    nlls[i] = nll
    
print('\n=========\n')
print(f'Loss (ave. nll): {nlls.mean().item()}')
```
