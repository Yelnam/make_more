```python
N = torch.zeros((27,27), dtype = torch.int32)
```

```python
chars = sorted(list(set(''.join(words))))
s_to_i = {s:i+1 for i, s in enumerate(chars)}
s_to_i['.'] = 0
s_to_i
```

```python
i_to_s = {i:s for s,i in s_to_i.items()}
i_to_s
```

```python
for word in words:
    tok = ['.'] + list(word) + ['.']
    for ch, ch2 in zip(tok, tok[1:]):
        ch_i = s_to_i[ch]
        ch2_i = s_to_i[ch2]
        N[ch_i, ch2_i] += 1
```

```python
plt.imshow(N)
```

```python
plt.figure(figsize = (16, 16))
plt.imshow(N, cmap = 'Blues')
for i in range(27):
    for j in range(27):
        ch_pair = i_to_s[i] + i_to_s[j]
        plt.text(j, i, ch_pair, ha = 'center', va = 'bottom', color = 'gray')
        plt.text(j, i, N[i, j].item(), ha = 'center', va = 'top', color = 'gray')
plt.axis('off');
```

```python
# seed to make random generations deterministic and repeatable
g = torch.Generator().manual_seed(2147483647)
p = torch.rand(3, generator = g)
p = p / p.sum()
p
```

```python
# multinomial based on above 'randomly' generated probabilities
torch.multinomial(p, num_samples = 20, replacement = True, generator = g)
```

```python
for i in range(15):
    out = []
    ix = 0
    while True:
        p = N[ix].float() #turn values in row ix (starting at row 0) into floats in order to perform calculations on them
        p = p /p.sum() # normalise row to produce probabilities summing to 1
        ix = torch.multinomial(p, num_samples = 1, replacement = True, generator = g).item() # get new ix value from a single sample on the row
        out.append(i_to_s[ix])
        if ix == 0:
            break
    print(''.join(out))
```

```python
# generate matrix of probabilities across rows (avoids inefficiency of renorming every run)
P = N.float()
row_sums = P.sum(1, keepdim = True) # 1 sums across rows (0 would sum columns)
# P = P / row_sums # divide the values in each cell of a row, by the row total (normalise the row)
P /= row_sums # commented out above, run /= to perform operation in place and avoid adding new memory
```

```python
P[26].sum()
```

```python
# run again, substituting the simpler P calculation
for i in range(15):
    out = []
    ix = 0
    while True:
        p = P[ix]
        ix = torch.multinomial(p, num_samples = 1, replacement = True, generator = g).item() # get new ix value from a single sample on the row
        out.append(i_to_s[ix])
        if ix == 0:
            break
    print(''.join(out))
```

```python
# run bigrams on early names again. this time include probabilities
log_likelihood = 0.0
n = 0
for word in words[20]:
    tok = ['.'] + list(word) + ['.']
    for ch, ch2 in zip(tok, tok[1:]):
        ch_i = s_to_i[ch]
        ch2_i = s_to_i[ch2]
        prob = P[ch_i, ch2_i]
        logprob = torch.log(prob)
        log_likelihood += logprob
        n += 1
        print(f'{ch}{ch2} \t%Prob: {prob*100:.2f} \tIndex: {(prob/(1/27)):.2f} \tLogProb: {logprob:.2f}')

nll = -log_likelihood
print(f'\nLog likelihood: {log_likelihood:.2f}, Neg log likelihood: {nll:.2f}, Ave. NLL: {nll/n:.7f}')
```
